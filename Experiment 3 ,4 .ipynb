```python
import pandas as pd
data = pd.read_csv("../Dataset_Ques_1_4.csv")
dataFrame = pd.DataFrame(data)
```

```python
dataFrame
```

```python
month = dataFrame['month_number']
facecream = dataFrame['facecream']
facewash = dataFrame['facewash']
toothpaste = dataFrame['toothpaste']
bathingsoap = dataFrame['bathingsoap']
moisturizer = dataFrame['moisturizer']
totalunits = dataFrame['total_units']
totalprofits = dataFrame['total_profit']
```

```python
import matplotlib.pyplot as plt

plt.plot(month, totalprofits, marker='o', linestyle='-')

plt.title("Company Profits units per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(month)
plt.yticks(totalprofits)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(month, total_units, color='red', marker='o', linestyle='--', linewidth=3, mfc='black', label='Profit data of last year')
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.xticks(month)
plt.yticks([10000, 20000, 30000, 40000, 50000])
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(month, totalunits, color='red', marker='o', linestyle='--', linewidth=3, mfc='black', label='Profit data of last year')
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.xticks(month)
plt.yticks([10000, 20000, 30000, 40000, 50000])
plt.legend(loc='lower right')
plt.show()
```

```python
plt.scatter(month, toothpaste, color='blue', alpha=0.75, label='Toothpaste sales data')
plt.title("Toothpaste sales data")
plt.xlabel("Month number")
plt.ylabel("Number of units sold")
plt.xticks(month)
plt.yticks([4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000])
plt.legend(loc='upper left')
plt.grid(True, linewidth= 1, linestyle='--')
plt.show()
```

```python
plt.plot(month, facecream, color='blue', linewidth=3, marker='o', label='Face cream sales data')
plt.plot(month, facewash, color='orange', linewidth=3, marker='o', label='Face wash sales data')
plt.plot(month, toothpaste, color='green', linewidth=3, marker='o', label='Toothpaste sales data')
plt.plot(month, bathingsoap, color='red', linewidth=3, marker='o', label='Bathing soap sales data')
plt.plot(month, moisturizer, color='brown', linewidth=3, marker='o', label='Moisturizer sales data')
plt.title("Sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.legend(loc='upper left')
plt.show()
```
